```python
# # imputing current
mode = df['Current'].value_counts().index.values[0]
df['Current'].fillna(mode, inplace=True)
print(df['Current'].isnull().value_counts())

# imputing Employee code

df["Employee Code"] = df["Employee Code"].astype("string")

df.loc[ df['Employee Code'] == "0", "Employee Code"] = np.nan
most_frequent_10 = df['Employee Code'].value_counts().head(10).index.values
freq_of_top_10 = df['Employee Code'].value_counts().head(10).values
empcode_probabilities= list(map(lambda value: (1/sum(freq_of_top_10))*value ,freq_of_top_10))
df['Employee Code']= df['Employee Code'].apply( lambda value: value if not pd.isnull(value) else np.random.choice(most_frequent_10, p= empcode_probabilities))


# imputing Production=> hyphen and name
df["Production"] = df["Production"].astype("string")
df.loc[ df['Production'] == "-", "Production"] = np.nan
df['Production'].isnull().value_counts()

df.loc[df['Production']=="#NAME?", "Production"] = "E15003253"

most_frequent_10 = df['Production'].value_counts().head(10).index.values
freq_of_top_10 = df['Production'].value_counts().head(10).values
production_probabilities= list(map(lambda value: (1/sum(freq_of_top_10))*value ,freq_of_top_10))
df['Production']= df['Production'].apply( lambda value: value if not pd.isnull(value) else np.random.choice(most_frequent_10, p= production_probabilities))

df['Production'].value_counts()


# imputing Order Operation No

df["Order Operation No"] = df["Order Operation No"].astype("string")

df.loc[ df['Order Operation No'] == "-", "Order Operation No"] = np.nan
most_frequent_10 = df['Order Operation No'].value_counts().head(10).index.values
freq_of_top_10 = df['Order Operation No'].value_counts().head(10).values
orderopno_probabilities= list(map(lambda value: (1/sum(freq_of_top_10))*value ,freq_of_top_10))
df['Order Operation No']= df['Order Operation No'].apply( lambda value: value if not pd.isnull(value) else np.random.choice(most_frequent_10, p= orderopno_probabilities))

df['Order Operation No'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
model = LabelEncoder()
df['Defect'] = model.fit_transform(df['Defect'].astype('str'))

y_label= df['Defect']
X_features= df[['Current', 'Humidity', 'Temperature', 'Flow', 'Job Temp','Voltage']]

folds=4 #75, 25 split
# random state- shuffling applied to the data before the split
# stratify- When it splits the label, it will make sure that the proportion of the training label is similar to the proportion of the testing label, for good accuracy

X_train, X_test, Y_train, Y_test= train_test_split(X_features, y_label, test_size=(1/folds), random_state= 42)
```
